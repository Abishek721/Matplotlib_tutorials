```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x= np.linspace(0,5,11)
y= x**2
```

```python
print(x)
print(y)
```

```python
#functional methods
plt.plot(x,y,"y",label="Xlabel")
plt.plot(y,x,"r",label="ylabel")
plt.title("title")
plt.legend(loc=0)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r")
plt.xlabel("Xlabel")
plt.ylabel("y_label")

```

```python
#object oriented methods

fig= plt.figure() 
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.1,0.1,0.4,0.4])

axes1.plot(x,y)
axes1.set_title("larger")
axes2.plot(y,x)
axes2.set_title("smaller")

```

```python
fig= plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y,color="b",lw=3,alpha=0.6,marker='o')
ax.text(3,0.5,x)
fig.show()
```
